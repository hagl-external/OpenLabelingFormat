```python
import os
import sys 
from random import randint
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
sys.path.append(os.path.abspath("../Loader"))
sys.path.append(os.path.abspath("../Loader/PyTorch"))
from OLFDetection2DDataset import OLFDetection2DDataset
```

```python
%matplotlib inline
```

```python
olf_dataset = OLFDetection2DDataset("../../Documents/instances_val2014_olf.json", ["image/png", "image/jpeg"], "../../../../CurrentDatasets/val2014")
```

```python
sample = olf_dataset[randint(0, len(olf_dataset) - 1)]
```

```python
plt.imshow(sample["img"])
```

```python
sample["box"]
```

```python
sample["cls"]
```

```python
assert len(sample["box"]) == len(sample["cls"])

fig,ax = plt.subplots(1)
ax.imshow(sample["img"])
for i, box in enumerate(sample["box"]):
    ax.add_patch(Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1], linewidth = 1, edgecolor = 'r', facecolor = 'none'))
    plt.text(box[0], box[1], sample["cls"][i], color = 'r')
plt.show()
```
